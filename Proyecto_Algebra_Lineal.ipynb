Proyecto Algebra Lineal 


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import imutils
```

```python
def Rot_Img(deg):
  theta=deg/100*np.pi
  c,s=np.cos(theta),np.sin(theta)
  return np.array([[c,-s],[s,c]])
```

Aca mostramos la imagen original

```python
# Que buen meme 
image = cv2.imread('imagen.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
ancho = image.shape[1] #columnas
alto = image.shape[0] # filas
```

```python
plt.imshow(image)
```

Traslacin de la imagen

```python
# Traslacin
M = np.float32([[1,0,10],[0,1,100]])
imageOut = cv2.warpAffine(image,M,(ancho,alto))
plt.imshow(imageOut)
```

Rotacin de la imagen

```python
#Rotacin
M = cv2.getRotationMatrix2D((ancho//2,alto//2),90,1)
imageOutR = cv2.warpAffine(image,M,(ancho,alto))
plt.imshow(imageOutR)
```

Escalonando la imagen

```python
#Escalando una imagen
imageOutE = cv2.resize(image,(200,200), interpolation = cv2.INTER_CUBIC)
plt.imshow(imageOutE)
```

Escalonado de la imagen por forma #2

```python
imageOutE = imutils.resize(image, width=200)
plt.imshow(imageOutE)
```
